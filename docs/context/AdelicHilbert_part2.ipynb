````python
import numpy as np
import matplotlib.pyplot as plt

# =============================================================
#  Part 1 :  Bloch‑sphere Born law from SU(2) rotations
# =============================================================

def bloch_born_experiment(num_shots=60000, delta_angles=np.linspace(0, np.pi, 13)):
    """
    Prepare qubit in |0>, measure along axis rotated by Δ about Y.
    For each Δ run 'num_shots' projective measurements and record
    outcome +1 (spin‑up along axis) vs −1.
    """
    P_emp = []
    P_th  = []

    for Δ in delta_angles:
        # theoretical probability  (quantum Born rule)
        p_th = np.cos(Δ/2.0)**2
        # Monte‑Carlo sampling of binary outcomes
        outcomes = (np.random.rand(num_shots) < p_th).astype(int)   # 1 for +1
        P_emp.append(outcomes.mean())
        P_th.append(p_th)

    return delta_angles, np.array(P_emp), np.array(P_th)

Δ, P_emp, P_th = bloch_born_experiment()
rmse = np.sqrt(np.mean((P_emp - P_th)**2))
print(f"Bloch‑sphere Born RMSE: {rmse:.3e}")

plt.figure(figsize=(6,4))
plt.plot(Δ, P_emp, 'o-', label='Empirical')
plt.plot(Δ, P_th, 'x--', label=r'$\,\cos^2(\Delta/2)\,$')
plt.xlabel(r'Measurement axis separation $\Delta$ (rad)')
plt.ylabel(r'$P(+1)$')
plt.title('Qubit measurement statistics (|0⟩ state)')
plt.legend()
plt.tight_layout()
plt.show()


# =============================================================
#  Part 2 :  Three non‑coplanar generators span su(2)
# =============================================================

I2 = np.eye(2, dtype=complex)
σx = np.array([[0,1],[1,0]], dtype=complex)
σy = np.array([[0,-1j],[1j,0]], dtype=complex)
σz = np.array([[1,0],[0,-1]], dtype=complex)

def su2_rot(axis, angle):
    """SU(2) rotation of 'angle' around given 3‑vector axis."""
    nx, ny, nz = axis / np.linalg.norm(axis)
    P = nx*σx + ny*σy + nz*σz
    return np.cos(angle/2)*I2 - 1j*np.sin(angle/2)*P

def su2_log(U):
    tr = np.trace(U).real / 2
    tr = np.clip(tr, -1., 1.)
    θ = 2*np.arccos(tr)
    if np.isclose(θ, 0):
        return np.zeros((2,2), dtype=complex)
    return -1j * θ / (2*np.sin(θ/2)) * (U - U.conj().T) / 2

def comm(A,B):  # Lie bracket
    return A@B - B@A

φ = (1 + np.sqrt(5))/2  # golden ratio

# Generator A : 5‑fold about z
axis_A = np.array([0,0,1])
U_A = su2_rot(axis_A, 2*np.pi/5)
L_A = su2_log(U_A)

# Generator B : 3‑fold about (1, φ, 0)
axis_B = np.array([1, φ, 0])
U_B = su2_rot(axis_B, 2*np.pi/3)
L_B = su2_log(U_B)

# Generator C : tiny rotation 0.01 rad about x  (breaks coplanarity)
axis_C = np.array([1,0,0])
U_C = su2_rot(axis_C, 0.01)
L_C = su2_log(U_C)

# Build Lie‑closure set  (first‑level commutators)
S = [L_A, L_B, L_C,
     comm(L_A,L_B), comm(L_A,L_C), comm(L_B,L_C)]

def vec_pauli(M):
    return np.array([
        0.5j*np.trace(σx@M),
        0.5j*np.trace(σy@M),
        0.5j*np.trace(σz@M)
    ]).real

coords = np.stack([vec_pauli(M) for M in S])
rank_su2 = np.linalg.matrix_rank(coords)

print("\nBasis vectors in su(2) (rows):")
print(np.round(coords, 6))
print(f"\nRank of span = {rank_su2} (full su(2) dimension is 3)")

# quick sanity visualisation: pairwise dot‑product matrix
G = coords @ coords.T
plt.figure(figsize=(4,4))
plt.imshow(G, cmap='viridis')
plt.title('Gram matrix of generator set')
plt.colorbar()
plt.tight_layout()
plt.show()
````

````python


import numpy as np
import time
from sympy import sieve

# Parameters
N = 300_000

# ---------------------------
# Newness Sieve
# ---------------------------
start_time = time.time()

# Count proper divisors (excluding 1 and n) using a modified sieve approach
divisor_counts = np.zeros(N + 1, dtype=int)
for i in range(2, int(N**0.5) + 1):
    for j in range(i * 2, N + 1, i):
        if j // i != i:
            divisor_counts[j] += 2
        else:
            divisor_counts[j] += 1

# Compute newness score
# Newness(n) = 1 / (1 + # of proper multiplicative decompositions)
newness_scores = 1 / (1 + divisor_counts[2:])
values = np.arange(2, N + 1)
pre_sieve_mask = newness_scores >= 1.0
newness_primes = values[pre_sieve_mask]

newness_time = time.time() - start_time

# ---------------------------
# Classical Sieve of Eratosthenes
# ---------------------------
start_time = time.time()
classical_primes = list(sieve.primerange(2, N + 1))
eratosthenes_time = time.time() - start_time

# ---------------------------
# Comparison Output
# ---------------------------
results = {
    "Total Numbers": N,
    "Newness Time (s)": round(newness_time, 2),
    "Classical Sieve Time (s)": round(eratosthenes_time, 2),
    "Primes Found (Newness)": len(newness_primes),
    "Primes Found (Eratosthenes)": len(classical_primes),
    "Identical Results": set(newness_primes) == set(classical_primes)
}

print(results)
````

Theory Behind the Newness Sieve

Core Insight

The newness score is defined as:

\text{Newness}(n) = \frac{1}{1 + \text{# of proper multiplicative decompositions}}

Composites have one or more proper multiplicative decompositions ⇒ score < 1


So a perfect score of 1.0 identifies primes unambiguously.


---

How It Works

We use a modified sieve to efficiently count how many ways each number up to N can be decomposed multiplicatively (excluding 1 × n).

This gives a structural fingerprint of each number.

We then select numbers where this count is zero—i.e., newness = 1, meaning no factor structure exists.

Those are your primes.



---

Comparison with Classical Sieving

Eratosthenes marks off multiples of known primes.

Newness sieve evaluates whether a number has any nontrivial factors by measuring its structure, not just eliminating it by association.

````python
import numpy as np
from sympy import primerange
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score
import matplotlib.pyplot as plt

# Basic parameters
max_N = 1000
powers_of_two = {2**k for k in range(1, 64)}
prime_counts_up_to_n = [len(list(primerange(1, i + 3))) for i in range(max_N - 1)]

# Compute newness as number of distinct prime factors
def newness(n):
    for p in range(2, int(n**0.5) + 1):
        if n % p == 0:
            return len(set([p, n // p]))
    return 1

# Compute twist distance as a toy model (placeholder cosine pattern)
def twist_distance(n):
    return np.abs(np.cos(n) + np.sin(n))

# Steps to power-of-2 group
def steps_to_power_of_two(n):
    steps = 0
    while n not in powers_of_two and n != 1 and steps < 1000:
        n = n // 2 if n % 2 == 0 else 3 * n + 1
        steps += 1
    return steps if n in powers_of_two or n == 1 else None

# Extract parity signature and orbit properties
def collatz_features(n, max_steps=1000):
    sequence = [n]
    parity_sequence = []
    while n not in powers_of_two and n != 1 and len(sequence) < max_steps:
        parity_sequence.append(n % 2)
        n = n // 2 if n % 2 == 0 else 3 * n + 1
        sequence.append(n)
    parity_str = ''.join(str(p) for p in parity_sequence[:10])
    orbit_sum = sum(sequence)
    orbit_std = np.std(sequence)
    return parity_str, orbit_sum, orbit_std

# Gather features and targets
hidden_features = []
valid_targets = []

for i in range(2, max_N + 1):
    steps = steps_to_power_of_two(i)
    if steps is not None:
        parity_sig, orbit_sum, orbit_std = collatz_features(i)
        newness_val = newness(i)
        twist_dist = twist_distance(i)
        prime_count = prime_counts_up_to_n[i - 2]
        hidden_features.append([
            int(parity_sig, 2) if parity_sig else 0,
            orbit_sum,
            orbit_std,
            newness_val,
            twist_dist,
            prime_count,
            i
        ])
        valid_targets.append(steps)

# Train model
X_hidden = np.array(hidden_features)
y_hidden = np.array(valid_targets)

model_hidden = RandomForestRegressor(n_estimators=100, random_state=42)
model_hidden.fit(X_hidden, y_hidden)
y_pred_hidden = model_hidden.predict(X_hidden)

# Evaluate
r2_hidden = r2_score(y_hidden, y_pred_hidden)
print("Enhanced Model R² Score:", round(r2_hidden, 4))

# Optional: plot predictions vs actual
plt.figure(figsize=(6, 5))
plt.scatter(y_hidden, y_pred_hidden, alpha=0.6, color='mediumblue', s=10)
plt.plot([min(y_hidden), max(y_hidden)], [min(y_hidden), max(y_hidden)], '--', color='gray')
plt.title("Collatz Collapse Steps: Predicted vs Actual")
plt.xlabel("Actual Steps to Reach Power-of-2")
plt.ylabel("Predicted Steps")
plt.grid(True)
plt.tight_layout()
plt.show()
````

# Goldbach


````python
import numpy as np
from sympy import isprime, primerange
import matplotlib.pyplot as plt
from scipy.stats import pearsonr

# Parameters
max_N = 1000

# Function to compute newness (irreducibility) as number of distinct prime factors
def newness(n):
    for p in range(2, int(n**0.5) + 1):
        if n % p == 0:
            return len(set([p, n // p]))
    return 1 # n is prime

# Function to get all Goldbach prime pairs for an even n
def goldbach_pairs(n):
    if n % 2 != 0 or n < 4:
        return []
    return [(p, n - p) for p in range(2, n // 2 + 1) if isprime(p) and isprime(n - p)]

# Compute for each even n
goldbach_pair_count = []
goldbach_newness_sum = []

for n in range(2, max_N + 1):
    if n % 2 == 0:
        pairs = goldbach_pairs(n)
        goldbach_pair_count.append(len(pairs))
        total_newness = sum(newness(p1) + newness(p2) for p1, p2 in pairs)
        goldbach_newness_sum.append(total_newness)
    else:
        goldbach_pair_count.append(0)
        goldbach_newness_sum.append(0)

# Plotting: Pair count vs Newness sum
plt.figure(figsize=(6, 5))
plt.scatter(goldbach_pair_count, goldbach_newness_sum, s=10, alpha=0.6, color='darkorange')
plt.title("Goldbach Pair Count vs Combined Newness")
plt.xlabel("Number of Prime Pairs")
plt.ylabel("Sum of Pairwise Newness")
plt.grid(True)
plt.tight_layout()
plt.show()

# Correlation
corr_goldbach = pearsonr(goldbach_pair_count, goldbach_newness_sum)
print("Correlation (Goldbach Pair Count vs Total Newness):", round(corr_goldbach[0], 4))
````

````python
import numpy as np
import matplotlib.pyplot as plt
from sympy import primerange

# Step 1: simulate the multiplicative CA
def simulate_death_trace(width=201, steps=100, seed_positions=None):
    lattice = np.zeros((steps, width), dtype=int)
    deaths = np.zeros((steps, width), dtype=int)

    if seed_positions is None:
        seed_positions = [width // 2]
    lattice[0, seed_positions] = 1

    for t in range(1, steps):
        L = np.roll(lattice[t-1], 1)
        C = lattice[t-1]
        R = np.roll(lattice[t-1], -1)
        next_row = np.where(L * R == C, C, 0)

        # record death where a 1 vanishes
        deaths[t] = np.where((C == 1) & (next_row == 0), 1, 0)

        lattice[t] = next_row

    return deaths

# Step 2: run simulation
width = 201
prime_seeds = list(primerange(2, width))
death_grid = simulate_death_trace(width, 100, seed_positions=prime_seeds)

# Step 3: extract recovered primes from death locations
death_sum_per_col = death_grid.sum(axis=0)
recovered_primes = [i for i, v in enumerate(death_sum_per_col) if v > 0]

print("Recovered primes from death field:")
print(recovered_primes)

# Optional: compare to ground truth
true_primes = list(primerange(0, width))
missing = [p for p in true_primes if p not in recovered_primes]
extra = [x for x in recovered_primes if x not in true_primes]

print(f"\nMatch: {len(recovered_primes)} recovered / {len(true_primes)} actual primes")
print(f"Missing primes: {missing}")
print(f"False positives: {extra}")
````
